```python
#### import gradio as gr
import joblib
import numpy as np
import gradio as gr
# =========================================================
# LOAD MODEL
# =========================================================
model = joblib.load("spotify_ml_final_model.pkl")
scaler = joblib.load("spotify_ml_scaler_final.pkl")

# =========================================================
# GENRE LIST
# =========================================================

genres = [
    "acoustic","afrobeat","alt-rock","alternative","ambient",
    "anime","black-metal","bluegrass","blues","brazil",
    "breakbeat","british","cantopop","chicago-house","children",
    "chill","classical","club","comedy","country",
    "dance","dancehall","death-metal","deep-house","detroit-techno",
    "disco","disney","drum-and-bass","dub","dubstep",
    "edm","electro","electronic","emo","folk",
    "forro","french","funk","garage","german",
    "gospel","goth","grindcore","groove","grunge",
    "guitar","happy","hard-rock","hardcore","hardstyle",
    "heavy-metal","hip-hop","honky-tonk","house","idm",
    "indian","indie","indie-pop","industrial","iranian",
    "j-dance","j-idol","j-pop","j-rock","jazz",
    "k-pop","kids","latin","latino","malay",
    "mandopop","metal","metalcore","minimal-techno","mpb",
    "new-age","opera","pagode","party","piano",
    "pop","pop-film","power-pop","progressive-house","psych-rock",
    "punk","punk-rock","r-n-b","reggae","reggaeton",
    "rock","rock-n-roll","rockabilly","romance","sad",
    "salsa","samba","sertanejo","show-tunes","singer-songwriter",
    "ska","sleep","songwriter","soul","spanish",
    "study","swedish","synth-pop","tango","techno",
    "trance","trip-hop","turkish","world-music"
]

# =========================================================
# PREDICTION FUNCTION
# =========================================================

def predict_song(
    danceability,
    energy,
    loudness,
    speechiness,
    acousticness,
    instrumentalness,
    liveness,
    valence,
    tempo,
    duration_ms,
    selected_genre
):

    try:

        # Convert genre to encoded value
        track_genre = genres.index(selected_genre)

        data = np.array([[
            danceability,
            energy,
            loudness,
            speechiness,
            acousticness,
            instrumentalness,
            liveness,
            valence,
            tempo,
            duration_ms,
            track_genre
        ]])
        #data scaler
        data_scaled = scaler.transform(data)
        
        # Prediction
        prediction = model.predict(data_scaled)[0]

        # Probability (if model supports it)
        try:
            probability = model.predict_proba(data_scaled)[0][1]
        except:
            probability = None

        if prediction == 1:
            result = "🎵 HIT SONG"
        else:
            result = "❌ NOT HIT SONG"

        if probability is not None:

            if probability >= 0.70:
                confidence = "HIGH"

            elif probability >= 0.40:
                confidence = "MEDIUM"

            else:
                confidence = "LOW"

            result += f"""

Genre: {selected_genre}

Hit Probability: {probability*100:.2f}%

Confidence: {confidence}
"""

        return result

    except Exception as e:
        return f"Error: {str(e)}"

# =========================================================
# GRADIO UI
# =========================================================

app = gr.Interface(
    fn=predict_song,

    inputs=[

        gr.Slider(
            minimum=0,
            maximum=1,
            value=0.5,
            step=0.01,
            label="Danceability"
        ),

        gr.Slider(
            minimum=0,
            maximum=1,
            value=0.5,
            step=0.01,
            label="Energy"
        ),

        gr.Slider(
            minimum=-60,
            maximum=0,
            value=-10,
            step=0.1,
            label="Loudness"
        ),

        gr.Slider(
            minimum=0,
            maximum=1,
            value=0.1,
            step=0.01,
            label="Speechiness"
        ),

        gr.Slider(
            minimum=0,
            maximum=1,
            value=0.5,
            step=0.01,
            label="Acousticness"
        ),

        gr.Slider(
            minimum=0,
            maximum=1,
            value=0.0,
            step=0.01,
            label="Instrumentalness"
        ),

        gr.Slider(
            minimum=0,
            maximum=1,
            value=0.2,
            step=0.01,
            label="Liveness"
        ),

        gr.Slider(
            minimum=0,
            maximum=1,
            value=0.5,
            step=0.01,
            label="Valence"
        ),

        gr.Slider(
            minimum=40,
            maximum=250,
            value=120,
            step=1,
            label="Tempo (BPM)"
        ),

        gr.Slider(
            minimum=30000,
            maximum=600000,
            value=180000,
            step=1000,
            label="Duration (ms)"
        ),

        gr.Dropdown(
            choices=genres,
            value="pop",
            label="Track Genre"
        )
    ],

    outputs=gr.Textbox(
        label="Prediction Result",
        lines=8
    ),

    title="🎵 Spotify Hit Song Prediction System",

    description="""
Enter song features and select a genre to predict whether the song is likely to be a Hit Song or a Not Hit Song.
"""
)

# =========================================================
# LAUNCH
# =========================================================


app.launch(
    share=True,
    css="""
    body {
        background: linear-gradient(
            135deg,
            #74ebd5,
            #ACB6E5
        ) !important;
    }

    .gradio-container {
        background: linear-gradient(
            135deg,
            #74ebd5,
            #ACB6E5
        ) !important;
    }

    /* Title */
    h1 {
        color: #003366 !important;
        font-weight: bold !important;
        text-align: center !important;
    }

    /* Description */
    .prose p {
        color: #003366 !important;
        font-size: 16px !important;
        text-align: center !important;
        font-weight: 500 !important;
    }

    footer {
        display: none !important;
    }
    """
)
```
